```python
import pandas as pd
import joblib
```

```python
# 모듈을 호출하여 신규 데이터를 예측해보기

# 신규 데이터 : 나이 : 35, 성별 : 남, 연소득 : 5000, 근속연수 : 8, 주거형태 : 자가, 신용점수 : 720,
# 기존대출건수 : 1, 연간카드사용액 : 2400, 부채비율 : 25.0, 대출신청액 : 3000, 대출목적 : 주택구입, 상환방식 : 원리금균등, 대출기간 : 36

# Module Load
pipe = joblib.load('../models/loan_pipeline.pkl')
encoders = joblib.load('../models/label_encoders.pkl')
cols_name = joblib.load('../models/feature_names.pkl')
```

```python
# Inference
sample_input = {
    "나이": 35,
    "성별": "남",
    "연소득": 5000,
    "근속연수": 8,
    "주거형태": "자가",
    "신용점수": 720,
    "기존대출건수": 1,
    "연간카드사용액": 2400,
    "부채비율": 25.0,
    "대출신청액": 3000,
    "대출목적": "주택구입",
    "상환방식": "원리금균등",
    "대출기간": 36
}

df_sample = pd.DataFrame([sample_input])
```

### Encoder Module : feature wise
> column transformer를 현업에서 활용한다고 함

```python
cat_cols = ['주거형태', '성별', '대출목적', '상환방식']

# Before
print('before : \n')
display(df_sample)

# Logics
for col in cat_cols:
    df_sample[col] = encoders[col].transform(df_sample[col])

# After
print('after : \n')
display(df_sample)
    
```

### Inference Module : Pipeline

```python
sample_pred = pipe.predict(df_sample)
print('예측한 값 : ', sample_pred)
```
